```python
#make a function to get data

def fetch_all_people():
    ''' 
    pulls in people data from swapi.dev and makes a df
    to run:
    people_data = fetch_all_people()
    people = pd.DataFrame(people_data)
    '''
    #set url
    url = "https://swapi.dev/api/people/"
    #make list
    people = []
    #make loop
    while url is not None:
        response = requests.get(url)
        data = response.json()
        people.extend(data["results"])
        url = data["next"]
    
    return people




```

```python
people_data = fetch_all_people()
people = pd.DataFrame(people_data)
```

## 2.Do the same thing, but for planets.

```python
def fetch_all_planets():
    ''' 
    pulls in planets data from swapi.dev and makes a df
    to run:
    planets_data = fetch_all_planets()
    planets = pd.DataFrame(planets_data)
    '''
    url = "https://swapi.dev/api/planets/"
    planets = []
    
    while url is not None:
        response = requests.get(url)
        data = response.json()
        planets.extend(data["results"])
        url = data["next"]
    
    return planets



```

```python
planets_data = fetch_all_planets()
planets = pd.DataFrame(planets_data)
```

```python
planets.shape
```

## 3. Extract the data for starships.

```python
def fetch_all_starships():
    ''' 
    pulls in starships data from swapi.dev and makes a df
    to run:
    starships_data = fetch_all_starships()
    starships = pd.DataFrame(starships_data)
    '''
    url = "https://swapi.dev/api/starships/"
    starships = []
    while url is not None:
        response = requests.get(url)
        data = response.json()
        starships.extend(data["results"])
        url = data["next"]
    return starships
```

```python
starships_data = fetch_all_starships()
starships = pd.DataFrame(starships_data)
```

```python
starships.shape
```

## 5. Combine the data from your three separate dataframes into one large dataframe.

```python
# Read the individual CSV files
people = pd.read_csv('people.csv')
planets = pd.read_csv('planets.csv')
starships = pd.read_csv('starships.csv')

# Concatenate the DataFrames
starwars = pd.concat([people, planets, starships], axis=0)

# Save the combined DataFrame to a new CSV file
starwars.to_csv('starwars.csv', index=False)
```

which has been rapidly expanding its renewable energy production in recent years. The data set includes country-wide totals of electricity consumption, wind power production, and solar power production for 2006-2017. You can get the data here: https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv

```python
# make function
def acquire_opsd_data():
    '''
    This function acquires the Open Power Systems Data for Germany (opsd) from
     "https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv"
     it returns an error message if failed
     # Usage ex:  opsd_data = acquire_opsd_data()
    '''
    url = "https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv"
    
    try:
        data = pd.read_csv(url)
        return data
    except Exception as e:
        print(f"Error occurred while acquiring data: {str(e)}")
        return None


```
